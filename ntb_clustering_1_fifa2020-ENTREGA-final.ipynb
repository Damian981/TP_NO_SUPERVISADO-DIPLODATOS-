**Análisis de skills en relación a las posiciones dentro del campo de juego**<br/>
A continuación se analizan los valores promedio de cada skill agrupando por la columna "team_position", y luego evaluando posiciones más generales: defensa, mediocampo, ataque. Para este análisis no se tendrán en cuenta los arqueros.

```python
# Se hace una copia del dataset tomando sólo los que tienen overall mayor a 70 y que no sean arqueros
df_to_copy = df[(df['team_position'] != 'GK') & (df['overall'] > 70)]

df_copy = df_to_copy.copy()
```

Previo al análisis, se filtra el dataframe con las 39 skills a utilizar.

```python
# Se hace un melt del dataframe a partir de las skills del dataframe copiado (no se tienen en cuenta arqueros) y se quitan los valores nulos
df_to_melt = df_copy[[
 'team_position', #0
 'shooting', #1
 'passing', #2
 'dribbling', #3
 'defending', #4
 'attacking_crossing', #11
 'attacking_finishing', #12
 'attacking_heading_accuracy', #13
 'attacking_short_passing', #14
 'attacking_volleys', #15
 'skill_dribbling', #16
 'skill_curve', #17
 'skill_fk_accuracy', #18
 'skill_long_passing', #19
 'skill_ball_control', #20
 'movement_acceleration', #21
 'movement_sprint_speed', #22
 'movement_agility', #23
 'movement_reactions', #24
 'movement_balance', #25
 'power_shot_power', #26
 'power_jumping', #27
 'power_stamina', #28
 'power_strength', #29
 'power_long_shots', #30
 'mentality_aggression', #31
 'mentality_interceptions', #32
 'mentality_positioning', #33
 'mentality_vision', #34
 'mentality_penalties', #35
 'mentality_composure', #36
 'defending_marking', #37
 'defending_standing_tackle', #38
 'defending_sliding_tackle' #39
]] 

df_melt = df_to_melt.melt(id_vars=["team_position"], 
        var_name="skill", 
        value_name="value")

df_melt.dropna(subset=["value"], inplace=True)
df_melt.head(10)
```

Calulamos el valor medio de cada skill agrupando por posición.

```python
df_melt = df_melt.groupby(['team_position', 'skill'], as_index=False).mean().astype({"value" : int})
```

Para cada posición, se le define una zona en el campo de juego según el último caracter de cada posición (B = Defensa, M / W = Mediocampo, S / T / F = Ataque)

```python
df_melt["zona"] = ""
for index, row in df_melt.iterrows():
    if row["team_position"].endswith("B"):
        df_melt.at[index, "zona"] = "DEF"
    elif row["team_position"].endswith("M") | row["team_position"].endswith("W"):
        df_melt.at[index, "zona"] = "MED"
    else: df_melt.at[index, "zona"] = "AT"
df_melt.head(4)
```

**Posiciones de defensa**

```python
# Análisis de puestos de defensa y skills
df_def = df_melt[(df_melt["zona"] == "DEF")].reset_index().pivot("team_position", "skill", "value")
plt.figure(figsize=(15,6))
sns.heatmap(df_def, annot=True)
```

**Posiciones de mediocampo**

```python
df_def = df_melt[(df_melt["zona"] == "MED")].reset_index().pivot("team_position", "skill", "value")
plt.figure(figsize=(15,6))
sns.heatmap(df_def, annot=True)
```

** Posiciones de ataque**

```python
df_def = df_melt[(df_melt["zona"] == "AT")].reset_index().pivot("team_position", "skill", "value")
plt.figure(figsize=(15,6))
sns.heatmap(df_def, annot=True)
```

**Análisis de lo observado en base a los 3 Heatmaps de posiciones y skills**<br/>


- Las posiciones de defensa tienen en promedio, valores más altos en skills de tipo "Defending" ("defending, defending_marking, defending_sliding_tackle, defending_stamina_tacke"), pero también se observan valores altos en algunos skills de movimiento y e forma aislada, en skills de potencia ("power_stamina, power_strength") y dribbling.
- Los mediocampistas se destacan en skills de movimiento (como "movement_acceleration" y "movement_agility"), aunque también muestran valores altos en promedio en dribbling, control de balón ("skill_ball_control") y pases cortos de ataque ("attacking_short_passing").
